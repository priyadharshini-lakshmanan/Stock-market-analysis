```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 30)) 
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
def categorize_corr(corr, pos_thresh=0.7, neg_thresh=-0.7):
    if corr >= pos_thresh and corr < 1.0:
        return 'Strong Positive'
    elif corr <= neg_thresh:
        return 'Strong Negative'
    elif abs(corr) < 0.2 and corr != 1.0:
        return 'No Correlation'
    else:
        return 'Moderate'

correlation_types = []
for i in correlation.columns:
    for j in correlation.columns:
        if i != j:
            corr_value = correlation.loc[i, j]
            correlation_types.append((i, j, corr_value, categorize_corr(corr_value)))

for pair in correlation_types:
    print(pair)
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/ACER/Desktop/stockwhole.csv', parse_dates=['date'])
df.columns = df.columns.str.lower()
df.set_index('date', inplace=True)
monthly_returns = df.groupby('ticker')['close'].resample('M').last().pct_change()
monthly_returns_df = monthly_returns.to_frame(name='monthlyreturns').reset_index()
monthly_returns_df.dropna(inplace=True)
monthly_returns_df
```

```python
N = 5
monthly_returns_df['gain'] = monthly_returns_df.groupby('date')['monthlyreturns'].rank(ascending=False, method='min')
monthly_returns_df['loss'] = monthly_returns_df.groupby('date')['monthlyreturns'].rank(ascending=True, method='min')
top_gainers_df = monthly_returns_df[monthly_returns_df['gain'] <= N].copy()
top_losers_df = monthly_returns_df[monthly_returns_df['loss'] <= N].copy()
top_gainers_df.sort_values(by=['date', 'monthlyreturns'], ascending=[True, False], inplace=True)
top_losers_df.sort_values(by=['date', 'monthlyreturns'], ascending=[True, True], inplace=True)
print("\n Top 5 Gainers Per Month")
top_gainers_df[['date', 'ticker', 'monthlyreturns']].head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv('C:/Users/ACER/Desktop/stockwhole.csv', parse_dates=['date'])
df.columns = df.columns.str.lower()

# Calculate monthly percent returns for each ticker
monthly = df.set_index('date').groupby('ticker')['close'].resample('M').last().pct_change()
monthly = monthly.dropna().reset_index()
monthly['month'] = monthly['date'].dt.strftime('%Y-%m')

# For each month, get top 5 gainers and losers
top_bottom = []
for m in monthly['month'].unique():
    group = monthly[monthly['month'] == m]
    gainers = group.nlargest(5, 'close')
    losers = group.nsmallest(5, 'close')
    top_bottom.append(gainers)
    top_bottom.append(losers)

final = pd.concat(top_bottom)
final['type'] = final['close'].apply(lambda x: 'Gainer' if x >= 0 else 'Loser')

# Plot: One vertical subplot per month (auto-handles non-12 months)
months = final['month'].unique()
fig, axes = plt.subplots(len(months), 1, figsize=(12, 4 * len(months)))
if len(months) == 1:
    axes = [axes]  # ensure axes is always iterable

for i, m in enumerate(months):
    subset = final[final['month'] == m]
    axes[i].bar(
        subset['ticker'],
        subset['close'],
        color=['green' if t == 'Gainer' else 'red' for t in subset['type']]
    )
    axes[i].set_title(f"Month: {m}")
    axes[i].set_ylabel('Monthly Return (%)')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_xlabel("Ticker")
plt.tight_layout()
plt.show()
```
