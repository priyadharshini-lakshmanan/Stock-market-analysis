```python
stockwhole['date'] = pd.to_datetime(stockwhole['date'])
df = stockwhole.sort_values(['Ticker', 'date'])

returns = []
for ticker, group in df.groupby('Ticker'):
    start = group.iloc[0]['close']
    end = group.iloc[-1]['close']
    yearly_return = (end / start) - 1
    returns.append({'Ticker': ticker, 'start close':start,'end close':end,'YearlyReturn': yearly_return})

    result = pd.DataFrame(returns)
   
```

```python
import matplotlib.dates as mdates
file_path = 'C:/Users/ACER/Desktop/cumulative_return.csv' 
df = pd.read_csv(file_path)
df['date'] = pd.to_datetime(df['date'])
df['cumulative_return'] = pd.to_numeric(df['cumulative_return'])
latest_date = df['date'].max()
latest_returns = df[df['date'] == latest_date]
top_5_tickers = latest_returns.sort_values(by='cumulative_return', ascending=False)['Ticker'].head(5).tolist()
df_top_5 = df[df['Ticker'].isin(top_5_tickers)]
ax = plt.gca() 

for ticker in top_5_tickers:
    ticker_data = df_top_5[df_top_5['Ticker'] == ticker]
    ax.plot(ticker_data['date'], ticker_data['cumulative_return'], linestyle='-', label=ticker)
    final_date = ticker_data['date'].iloc[-1]
    final_value = ticker_data['cumulative_return'].iloc[-1]
    ax.annotate(
        f'{final_value:.5f}',  
        xy=(final_date, final_value),  
        xytext=(5, 0),  
        textcoords="offset points", 
        ha='left',      
        va='center',    
        fontsize=9,     
        color='black'   
    )

plt.title('Top 5 Performing Stocks Cumulative Return Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend(title='Ticker')
plt.grid(True)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) 
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/ACER/Desktop/stockwhole.csv', parse_dates=['date'])
df.columns = df.columns.str.lower()
df.set_index('date', inplace=True)
monthly_returns = df.groupby('ticker')['close'].resample('M').last().pct_change()
monthly_returns_df = monthly_returns.to_frame(name='monthlyreturns').reset_index()
monthly_returns_df.dropna(inplace=True)
monthly_returns_df
```

```python
N = 5
monthly_returns_df['gain'] = monthly_returns_df.groupby('date')['monthlyreturns'].rank(ascending=False, method='min')
monthly_returns_df['loss'] = monthly_returns_df.groupby('date')['monthlyreturns'].rank(ascending=True, method='min')
top_gainers_df = monthly_returns_df[monthly_returns_df['gain'] <= N].copy()
top_losers_df = monthly_returns_df[monthly_returns_df['loss'] <= N].copy()
top_gainers_df.sort_values(by=['date', 'monthlyreturns'], ascending=[True, False], inplace=True)
top_losers_df.sort_values(by=['date', 'monthlyreturns'], ascending=[True, True], inplace=True)
print("\n Top 5 Gainers Per Month")
top_gainers_df[['date', 'ticker', 'monthlyreturns']].head(5)
```

```python
print("\nTop 5 Losers Per Month")
top_losers_df[['date', 'ticker', 'monthlyreturns']].head(5)
```

```python
import plotly.express as px
N = 5
top_bottom_filtered_df = monthly_returns_df[(monthly_returns_df['rank_gain'] <= N) | (monthly_returns_df['rank_loss'] <= N)].copy()
top_bottom_filtered_df['Type'] = top_bottom_filtered_df['monthlyreturns'].apply(lambda x: 'Gainer' if x >= 0 else 'Loser')
top_bottom_filtered_df['Month'] = top_bottom_filtered_df['date'].dt.strftime('%Y-%m')
fig = px.bar(
    top_bottom_filtered_df,
    x='ticker',
    y='monthlyreturns',
    color='Type',
    facet_col='Month',
    facet_col_wrap=3, 
    title='Top 5 Monthly Gainers and Losers Dashboard',
    color_discrete_map={'Gainer': 'green', 'Loser': 'red'},
    height=800,
    width=1000
)
fig.update_yaxes(tickformat=".1%")
fig.show()
```
